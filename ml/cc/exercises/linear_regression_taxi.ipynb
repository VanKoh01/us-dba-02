```python
# Análisis de Regresión Lineal: Ruta de Ciclismo
# Relación entre kilómetros recorridos y puntos visitados

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def main():
    # 1. Datos proporcionados
    # X = Kilómetros recorridos (variable independiente)
    # y = Puntos visitados (variable dependiente)
    X = np.array([[22.5], [13.0], [31.8], [20.2], [7.0], [9.9], [15.0]])
    y = np.array([8, 5, 16, 4, 8, 11, 20])

    # 2. Visualización inicial de los datos
    print("\n=== Visualización de Datos ===")
    plt.figure(figsize=(10, 5))
    plt.scatter(X, y, color='blue', s=100, alpha=0.7)
    plt.title('Relación entre Km Recorridos y Puntos Visitados', fontsize=14)
    plt.xlabel('Kilómetros Recorridos', fontsize=12)
    plt.ylabel('Puntos Visitados', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # 3. Creación y entrenamiento del modelo de regresión lineal
    modelo = LinearRegression()
    modelo.fit(X, y)

    # 4. Obtención de parámetros del modelo
    a = modelo.coef_[0]  # Pendiente (coeficiente)
    b = modelo.intercept_  # Intercepto

    print("\n=== Resultados del Modelo ===")
    print(f"Ecuación de la recta: y = {a:.4f}*X + {b:.4f}")
    print(f"Interpretación: Por cada km adicional, se visitan {a:.2f} puntos en promedio")
    print(f"Intercepto: {b:.2f} puntos visitados cuando no se recorren km (interpretación teórica)")

    # 5. Predicciones y evaluación del modelo
    y_pred = modelo.predict(X)
    r2 = r2_score(y, y_pred)

    print("\n=== Evaluación del Modelo ===")
    print(f"Coeficiente de determinación R²: {r2:.4f}")
    print("R² indica qué porcentaje de la variación en puntos visitados es explicada por los km recorridos")
    print(f"En este caso, el modelo explica el {r2*100:.1f}% de la variabilidad")

    # 6. Visualización de la regresión
    plt.figure(figsize=(10, 5))

    # Datos reales
    plt.scatter(X, y, color='blue', s=100, alpha=0.7, label='Datos reales')

    # Línea de regresión
    plt.plot(X, y_pred, color='red', linewidth=2, label='Regresión lineal')

    # Personalización del gráfico
    plt.title('Regresión Lineal: Km Recorridos vs Puntos Visitados', fontsize=14)
    plt.xlabel('Kilómetros Recorridos', fontsize=12)
    plt.ylabel('Puntos Visitados', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)

    # Mostrar ecuación en el gráfico
    equation = f'y = {a:.2f}*X + {b:.2f}\nR² = {r2:.2f}'
    plt.text(0.5, 0.9, equation, fontsize=12, transform=plt.gca().transAxes,
             bbox=dict(facecolor='white', alpha=0.8))

    plt.show()

    # 7. Análisis adicional
    print("\n=== Análisis Adicional ===")
    print("Observaciones sobre los datos:")
    print("- No hay una relación lineal fuerte (R² bajo)")
    print("- Los puntos más altos no siempre corresponden a mayores distancias")
    print("- Podrían influir otros factores como dificultad del terreno o interés turístico")
    print("\nRecomendación: Considerar añadir más variables explicativas para mejorar el modelo")

if __name__ == "__main__":
    main()
```
